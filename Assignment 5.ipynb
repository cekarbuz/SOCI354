```python
# Because I encountered SO MANY problems with 'dimension
# mismatch'es, I took some cautionary measures

usdata_list = list(us_data["GISJOIN"])
usshape_list = list(us_shapefile["GISJOIN"])
data_del= []
shape_del =[]
for x in usdata_list:
    if x not in usshape_list:
        data_del.append(x)
for x in usshape_list:
    if x not in usdata_list:
        shape_del.append(x)
        
us_data = us_data[~us_data['GISJOIN'].isin(data_del)]
us_shapefile = us_shapefile[~us_shapefile['GISJOIN'].isin(shape_del)]




usdata_list = list(us_data["GISJOIN"])
usshape_list = list(us_shapefile["GISJOIN"])

for x in usdata_list:
    if x not in usshape_list:
        print(x)

for x in usshape_list:
    if x not in usdata_list:
        print(x)
```

# Step 2: Spatial matrix

```python
# Create the spatial weight matrix using the Queen method

import libpysal
from libpysal.weights import Queen

wq = libpysal.weights.Queen.from_dataframe(merged_data_us)
wq.transform = 'r'
wq
```

```python
merged_data_us['AQP6E001_splag'] = libpysal.weights.lag_spatial(wq, merged_data_us['AQP6E001'])

merged_data_us['AQP6E001_splag'], merged_data_us['AQP6E001']
```
